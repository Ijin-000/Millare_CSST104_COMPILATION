**1. CREATING DATASET**


```python
import pandas as pd

data = {
    'student_id': range(1, 16),
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'test_date': pd.date_range(start = '2024-02-12', periods = 15 ),
    'study_hours': [5,6,7,8,6,9,8,9,6,4,6,7,8,5,3],
    'test_score': [87,88,89,93,87,95,93,92,91,85,89,88,87,86,83],
    'attendance': ['Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present'],

}

df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index=False)
```

**2. DATA LOADING AND INITIAL EXAMINATION**

```python
df = pd.read_csv('student_performance.csv')

print(df.head(15))
print(df.isnull().sum())
```

**3. EXPLORATORY DATA ANALYSIS**

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score', 'study_hours']].describe()

# Histogram
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sne.histplot(df['test_score'],bins=15, kde=True)
plt.title('Distribution of Test Score')


# Scatter Plot
plt.subplot(1,2,2)
sne.histplot(df['study_hours'],bins=15, kde=True)
plt.title('Distribution of Study Hours')


plt.figure(figsize=(8,6))
sne.scatterplot(data = df, x= 'study_hours', y = 'test_score')
plt.title('Scatter Plot of Study Hours vs. Test Scores')
```

**4. DATA VISUALIZATION**

```python
# Gender
plt.figure(figsize=(6,4))
sne.countplot(data=df, x = 'gender')
plt.title('Gender Distribution')

plt.figure(figsize=(10,8))
sne.lineplot(data = df, x = 'test_date', y ='test_score')
plt.title('Trends in Test Score Over Times')
plt.xticks(rotation = 45)

correlation_matrix = df[['study_hours', 'test_score','attendance']].corr()
plt.figure(figsize=(8,6))
sne.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
```
