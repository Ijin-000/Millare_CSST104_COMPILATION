**1. Importing Libraries**

```python
import pandas as pd # Data Manipulation
import numpy as np # Mathematical Operation
import matplotlib.pyplot as plt # Data Visualization
from sklearn.linear_model import LinearRegression # Data Training/Data Modeling
```

**2. Import CSC File:**

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
```

**3. Create DataFrame**

```python
prev_examscore = data['Previous_Exam_Scores']
grades = data['Grades']
```

**3. Visualizing Data**

```python
plt.scatter(prev_examscore,grades)
plt.xlabel('Independent Variable (prev_examscore)')
plt.ylabel('Dependent Variable (grades)')
plt.title('Scatter Plot of Data')
plt.show()
```

**4. PREPARING DATA FOR MODELING:**

```python
prev_examscore = data[['Previous_Exam_Scores']]
grades = data[['Grades']]
```

**5. Fitting the Linear Regression Model:**

```python
model = LinearRegression()
model.fit(prev_examscore,grades)
```

**6. Getting Model Coefficients:**

```python
slope = model.coef_[0]
intercept = model.intercept_
print('Slope:', slope)
print('Intercept:', intercept)
```

**7. Visualizing the Linear Regression Line:**

```python
plt.scatter(prev_examscore,grades)
plt.plot(prev_examscore, model.predict(prev_examscore), color= 'orange')
plt.xlabel('Independent Variable (prev_examscore)')
plt.ylabel('Dependent Variable (grades)')
plt.title('Linear Regression')
plt.show()
```
